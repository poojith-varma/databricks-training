Select Multiple Columns

```python
%sql

SELECT
    name,
    department,
    salary
FROM employees;
```

Using col()

```python
%sql

SELECT
    name,
    city,
    experience
FROM employees;
```

```python
display(
    df.select(
        col("name"),
        col("city"),
        col("experience")
    )
)
```

Add Computed Column

```python
%sql

SELECT
    *,
    salary + 5000 AS salary_after_bonus
FROM employees;
```

```python
display(
    df.withColumn(
        "salary_after_bonus",
        col("salary") + 5000
    )
)
```

Add Constant Column

```python
%sql

SELECT
    *,
    'India' AS country
FROM employees;
```

```python
display(
    df.withColumn(
        "country",
        lit("India")
    )
)
```

Alias

```python
%sql

SELECT
    name AS employee_name,
    salary AS monthly_salary
FROM employees;
```

```python
display(
    df.select(
        col("name").alias("Employee Name"),
        col("salary").alias("Monthly Salary")
    )
)
```

Rename Column

```python
%sql

SELECT
    employee_id,
    name,
    age,
    department AS dept,
    salary,
    city,
    experience,
    status
FROM employees;
```

```python
display(
    df.withColumnRenamed(
        "department",
        "dept"
    )
)
```

**Filtering Data**

Salary > 70000

```python
%sql

SELECT *
FROM employees
WHERE salary > 70000;
```

```python
display(
    df.filter(
        col("salary") > 70000
    )
)
```

Age > 30

```python
%sql

SELECT *
FROM employees
WHERE age > 30;
```

```python
display(
    df.where(
        "age > 30"
    )
)
```

AND

```python
%sql

SELECT *
FROM employees
WHERE salary > 70000
AND department = 'Finance';
```

```python
display(
    df.filter(
        "salary > 70000 AND department='Finance'"
    )
)
```

OR

```python
%sql

SELECT *
FROM employees
WHERE department = 'HR'
OR department = 'Research';
```

```python
display(
    df.filter(
        (col("department") == "HR") |
        (col("department") == "Research")
    )
)
```

NOT

```python
%sql

SELECT *
FROM employees
WHERE NOT status = 'Inactive';
```

```python
display(
    df.filter(
        ~(col("status") == "Inactive")
    )
)
```

NULL Values(AGE)

```python
%sql

SELECT *
FROM employees
WHERE age IS NULL;
```

```python
display(
    df.filter(
        col("age").isNull()
    )
)
```

NOT NULL Values(DEPARTMENT)

```python
%sql

SELECT *
FROM employees
WHERE department IS NOT NULL;
```

```python
display(
    df.filter(
        col("department").isNotNull()
    )
)
```
